## OO Method from Scratch

```python
fig,ax=plt.subplots(figsize=(8,4))
scatter_heart=ax.scatter(x=heart_over_60["age"],
                        y=heart_over_60["chol"],
                        c=heart_over_60["sex"])
ax.set(title="Heart Paitents",
       xlabel="Age",
       ylabel="Cholestrol");
ax.legend(*scatter_heart.legend_elements(), title="Sex")
# Adding a horizontal line to show mean
ax.axhline(heart_over_60["chol"].mean(),
          linestyle="--");
```

## Creating Subplots using OO method

```python
fig,(ax1,ax2)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10))
plot1=ax1.scatter(x=heart_over_60["chol"],
                 y=heart_over_60["age"],
                 c=heart_over_60["target"]);
ax1.set(title="Heart Cholestrol Levels",
       xlabel="Cholestrol",
       ylabel="Age")
ax1.legend(*plot1.legend_elements(), title="Target");
ax1.axhline(heart_over_60["age"].mean(),
         linestyle="--");
plot2=ax2.scatter(x=heart_over_60["age"],
                  y=heart_over_60["cp"],
                  c=heart_over_60["sex"])
ax2.set(title="CP W.R.T Age",
       xlabel="Age",
       ylabel="CP")
ax2.legend(*plot1.legend_elements(), title="Sex");
```
